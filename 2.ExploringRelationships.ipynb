```python
# Create a subset excluding the "Date" and "No" columns

sub = df[['cumulative_cases',
       'new_cases', 'cumulative_deaths', 'new_deaths', 'total_vaccinations',
       'people_vaccinated', 'people_fully_vaccinated', 'population']]
```

```python
sub.head(10)
```

```python
# Create a correlation matrix using pandas
sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

### Create a correlation heatmap using seaborn:

```python
# Create a subplot with matplotlib. What it does is create an empty output, but this is okay—it's just to show the subplots.

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
## The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

```python
corr.figure
```

##### Purple equates to 0, or no relationship between the variables, dark purple and black equate to negative correlations, and orange and light orange equate to positive correlations. As per this we can see that these three are the strongest relationship between variable we have discovered using this heat map

##### 1. Cumulative_cases and Cumulative_Deaths ( Coefficient 0.93)
##### 2. Cumulative_cases and people_fully_vaccinated ( Coefficient 0.63)
##### 3. Cumulative_deaths and people_vaccinated (Coefficient 0.63)

#### Scatterplots

```python
# Create a scatterplot for the "cumulative_cases" and "cumulative_deaths" columns in seaborn
sns.lmplot(x = 'cumulative_cases', y = 'cumulative_deaths', data = sub)
```

##### The trend of this scatter plot shows that there is a very strong relationship between cumulative cases and cumulative deaths. Wherever more cases are coming , those countries are facing high mortality

```python
# Create a scatterplot for the "cumulative_deaths" and "people_vaccinated" columns in seaborn
sns.lmplot(x = 'cumulative_deaths', y = 'people_vaccinated', data = sub)
```

#### This scatter plot and its trend line shows the strong relationship between cumulative deaths and people_vaccinated , but we can clearly see the trend line is much flatter and  that there are many outliers, which could state that deaths are not that consistent in the countries where the vaccination rate is progressing

```python
# Create a scatterplot for the "cumulative_cases" and "people_vaccinated" columns in seaborn
sns.lmplot(x = 'cumulative_cases', y = 'people_fully_vaccinated', data = sub)
```

##### This scatter plot and its trend line shows a strong relationship between people_fully_vaccinated and cumulative_cases. But the trend line here is not very steep, more on a flat side alongwith many outlier datapoints, which could state that where the people are fully vaccinated , number of cases are reducing.

#### Pair Plots

```python
# Keep only the variables you want to use in the pair plot

sub_2 = sub[['cumulative_cases', 'cumulative_deaths','total_vaccinations' ,'people_vaccinated','population']]
```

```python
# Create a pair plot 

g = sns.pairplot(sub_2)
```

##### Analysis of Pair Plots

##### When we crossed the two variables - Cumulative cases and cumulative deaths, we can see the strongest relatiosnhip, more cases and high mortaility. There is one another variable people_vaccinated, when we crossed this variable with cumulative_cases, we found out that initially when there are less people vaccinated , the cases were almost consistent but as the vaccination rate increased , we can clearly see the slowdown in number of cases. Similarly when we crossed people_vaccinated with cumulative deaths, almost similar visualization came out, where the death rate is slowing down with increased rate of vaccination.

#### Creating a categorical variable that splits the "cumulative_deaths" column into categories

```python
## Use a histogram to visualize the distribution of the variable -cumulative_deaths
sns.displot(df['cumulative_deaths'], bins = 80)
```

```python
df.loc[df['cumulative_deaths'] < 20000, 'mortality range'] = 'Low mortality'
```

```python
df.loc[(df['cumulative_deaths'] >= 20000) & (df['cumulative_deaths'] < 40000), 'mortality range'] = 'Medium mortality'
```

```python
df.loc[df['cumulative_deaths'] >= 40000, 'mortality range'] = 'High mortality'
```

```python
df['mortality range'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="location", y="cumulative_deaths", hue="mortality range", data=df)
```
